### Questão 3:

image.png

```python
## Importando bibliotecas
import math
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

# Dados da questão
distancia_1 = 40
velocidade_1 = 30

distancia_2 = 40
velocidade_2 = 60

# Letra a)
delta_s = distancia_1 + distancia_2 # Calculando Delta s

delta_t1 = distancia_1 / velocidade_1 # Calculando Delta t1
delta_t2 = distancia_2 / velocidade_2 # Calculando Delta t2
delta_t = delta_t1 + delta_t2 # Calculando Delta t

velocidade_media = delta_s / delta_t # Calculando a velocidade média

print("a) {} km/h".format(velocidade_media)) # Imprimindo resposta

# Letra b)

velocidade_escalar_media = velocidade_media # Atribuindo o valor da velocidade escalar média

print("b) Como não houve nenhuma mudança na direção\n   e sentido, então o módulo da velocidade\n   média é igua a velocidade escalar média:\n   {} km/h".format(velocidade_escalar_media)) # Imprimindo resposta

# Letra c)
## Calculando intervalos de tempo
tempo_1 = distancia_1 / velocidade_1
tempo_2 = distancia_2 / velocidade_2

## Definindo os vetores das distâncias percorridas
eixo_x = np.array([0, tempo_1, tempo_1 + tempo_2])
eixo_y = np.array([0, distancia_1, distancia_1 + distancia_2])

## Definindo o vetor da velocidade média
velocidade_media_x = np.array([0, tempo_1 + tempo_2])
velocidade_media_y = np.array([0, distancia_1 + distancia_2])

## Plotando o gráfico
plt.plot(eixo_x, eixo_y, label = "Trajeto")
plt.plot(velocidade_media_x, velocidade_media_y, linestyle="dashed", label = "Velocidade média")

plt.vlines(eixo_x, 0, eixo_y, linestyle="dotted")
plt.hlines(eixo_y, 0, eixo_x, linestyle="dotted")

plt.xlabel("t(h)")
plt.ylabel("x(km)")
plt.legend()

plt.xlim(0,None)
plt.ylim(0,None)

plt.grid(True)

## Imprimindo resposta
print("c)")
plt.show()
```

```python
# Dados da questão
def x(t): # Função da equação x
  return(3 * t) - ( 4 * (t ** 2)) + (t ** 3)

# Letra a)
posicao = x(1) # Calculando posição com t = 1s
print("a) {}".format(posicao)) # Imprimindo resposta

# Letra b)
posicao = x(2) # Calculando posição com t = 2s
print("b) {}".format(posicao)) # Imprimindo resposta

# Letra c)
posicao = x(3) # Calculando posição com t = 3s
print("c) {}".format(posicao)) # Imprimindo resposta

# Letra d)
posicao = x(4) # Calculando posição com t = 4s
print("d) {}".format(posicao)) # Imprimindo resposta

# Letra e)
deslocamento = x(4) - x(0) # Calculando o deslocamento
print("e) {} m".format(deslocamento))

# Letra f)
velocidade_media = (x(4) - x(2)) / (4 - 2)
print("f) {} m/s".format(velocidade_media))

# Letra g)
def plotar_grafico(a, b, c, d):
    eixo_x = np.arange(0, 4.01, 0.05)
    eixo_y = [(a * i**3 + b * i**2 + c * i + d) for i in eixo_x]
    plt.plot(eixo_x, eixo_y)
    plt.xlabel("t(s)")
    plt.ylabel("x(m)")
    plt.grid(True)
    plt.show()

print("g)")
plotar_grafico(1, -4, 3, 0)
```

```python
# Letra a)
## Derivando a função
t = sp.Symbol("t")
funcao_velocidade = sp.diff(4 - (12 * t) + (3 * (t ** 2)))

print("a) {} m/s".format(funcao_velocidade.subs(t, 1))) # Imprimindo resposta

# Letra b)
print("b) Como a velocidade é negativa, o movimento é no sentido negativo.") # Imprimindo resposta

# Letra c)
print("c) {} m/s".format(math.fabs(funcao_velocidade.subs(t, 1))))

# Letra d)
print("d) Com t = 1.1 a velocidade escalar é de {:.2} logo a velocidade está diminuindo.".format(funcao_velocidade.subs(t, 1.1)))

# Letra e)
print("e) Existe, quando t = {}".format(sp.solve(funcao_velocidade, t)[0]))

# Letra f)

# 100 linearly spaced numbers
x = np.arange(-0.1, 5.1, 0.05)

# the function, which is y = x^2 here
y = 6 * x - 12

# plot the function
plt.xlabel("t(s)")
plt.ylabel("x(m)")
plt.grid(True)
plt.plot(x,y, 'r')

print("f) Não existe, como pode ser visto no gráfico")


# show the plot
plt.show()
```
